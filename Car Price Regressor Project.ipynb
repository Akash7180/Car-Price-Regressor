# User Defined score card

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_percentage_error

# Initialize score_card DataFrame outside the function to avoid overwriting
score_card = pd.DataFrame(columns=['model name', 'r2_score', 'RMSE_train', 'RMSE_test', 'MAPE'])

def update_score_card(model_name, model, xtrain, xtest, ytrain, ytest):
    global score_card  # To modify the global score_card within the function
    model.fit(xtrain, ytrain)
    ypred_train = model.predict(xtrain)
    ypred_test = model.predict(xtest)
    new_data = pd.DataFrame({
        'model name': [model_name],
        'r2_score': [r2_score(ytest, ypred_test)],
        'RMSE_train': [np.sqrt(mean_squared_error(ytrain, ypred_train))],
        'RMSE_test': [np.sqrt(mean_squared_error(ytest, ypred_test))],
        'MAPE': [mean_absolute_percentage_error(ytest, ypred_test)]
    })
    score_card = pd.concat([score_card, new_data], ignore_index=True)
    return score_card
```

```python
x = dff[['wheelbase',
 'enginesize',
 'stroke',
 'peakrpm',
 'CompanyName',
 'aspiration_turbo',
 'enginelocation_rear']]
y = dff['price']
xtrain,xtest,ytrain,ytest =  train_test_split(x,y,test_size=0.3,random_state=10)
lr = LinearRegression()
update_score_card("LinearRegression",lr,xtrain,xtest,ytrain,ytest)
```

### MODEL 6 : Ridge Regression Hyperparameter Tuning

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

# Hyperparameters to tune
param_grid_ridge = {'alpha': [0.01, 0.1, 1, 10, 100]}

# GridSearchCV
ridge = Ridge()
grid_ridge = GridSearchCV(ridge, param_grid_ridge, scoring='r2', cv=5)
grid_ridge.fit(xtrain, ytrain)

# Best estimator
best_ridge = grid_ridge.best_estimator_

# Update score card
print("Best Ridge parameters:", grid_ridge.best_params_)
update_score_card("tuned_Ridge", best_ridge, xtrain, xtest, ytrain, ytest)
```

### How well those variables describe the price of a car.

Wheelbase: Longer wheelbases are often associated with more stable and comfortable rides, typically found in luxury cars, which can fetch higher prices.

Engine Size (Enginesize): Larger engines usually mean more power and better performance, leading to higher costs. However, they may also result in higher fuel consumption.

Stroke: This variable impacts engine performance and torque. Cars with optimal stroke values can offer better driving experiences, which can influence their market value.

Peak RPM (Peakrpm): Higher peak RPMs suggest a performance-oriented engine, appealing to buyers looking for speed and agility, and thus can drive up the price.

Company Name (CompanyName): Brand reputation heavily influences car prices. Premium brands like BMW, Mercedes, and Audi often command higher prices due to their perceived quality and status.

Aspiration Turbo (Aspiration_turbo): Turbocharged engines generally provide better performance and fuel efficiency, making cars with this feature more desirable and expensive.

Engine Location (Enginelocation_rear): Rear-engine layouts are often found in high-performance sports cars, which are typically priced higher due to their unique design and handling characteristics.
